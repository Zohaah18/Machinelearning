```python
#Question 2:
import pandas as pd
import numpy as np
from collections import Counter

train_df = pd.read_csv("occupancy_train.csv")
test_df = pd.read_csv("occupancy_test.csv")

X_train = train_df.drop(columns=["date","Occupancy"]).values
y_train = train_df["Occupancy"].values
X_test = test_df.drop(columns=["date","Occupancy"]).values
y_test = test_df["Occupancy"].values

def chi_square_distance(x,y):
    return np.sum(((x-y)**2)/(x+y+1e-10))

class KNNClassifier:
    def __init__(self,k=5): self.k=k
    def fit(self,X,y): self.X_train,self.y_train=X,y
    def predict_one(self,x):
        distances=[chi_square_distance(x,xt) for xt in self.X_train]
        k_idx=np.argsort(distances)[:self.k]
        k_labels=[self.y_train[i] for i in k_idx]
        return Counter(k_labels).most_common(1)[0][0]
    def predict(self,X): return [self.predict_one(x) for x in X]

knn=KNNClassifier(k=5)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)

accuracy=np.mean(np.array(y_pred)==y_test)
print("Accuracy:",accuracy)

def confusion_matrix(y_true,y_pred):
    labels=sorted(list(set(y_true)))
    m=np.zeros((len(labels),len(labels)),dtype=int)
    for t,p in zip(y_true,y_pred): m[t][p]+=1
    return m

print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
```

```python
#Question 3:
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

DATA = "cancer patient data sets.csv"
OUTDIR = "Lab 03 KNN.pdf"
RANDOM_STATE = 0
TOP_FEATURES = 12
KS = [1, 3, 5, 7, 9]
METRICS = {"euclidean": 2, "manhattan": 1, "minkowski_p3": 3}


df = pd.read_csv(DATA)
target = "Label" if "Label" in df.columns else ("Level" if "Level" in df.columns else df.columns[-1])
print("Using target:", target)
print("Class counts:\n", df[target].value_counts())


for idc in ["index", "Patient Id", "Patient_ID", "Id"]:
    if idc in df.columns:
        df = df.drop(columns=[idc])
dups = df.duplicated().sum()
if dups:
    df = df.drop_duplicates().reset_index(drop=True)
if df.isnull().sum().sum() > 0:
    for c in df.columns:
        if df[c].isnull().any():
            if pd.api.types.is_numeric_dtype(df[c]):
                df[c] = df[c].fillna(df[c].median())
            else:
                m = df[c].mode(dropna=True)
                df[c] = df[c].fillna(m[0] if len(m)>0 else "missing")

X = df.drop(columns=[target]).copy()
y_raw = df[target].copy()
for col in X.select_dtypes(exclude=[np.number]).columns:
    if X[col].nunique() <= 20:
        X[col] = LabelEncoder().fit_transform(X[col].astype(str))
    else:
        X[col] = X[col].astype(str).map(lambda s: abs(hash(s)) % (10**8))
le = LabelEncoder(); y = le.fit_transform(y_raw)

corr = pd.concat([X, pd.Series(y, name="__t__")], axis=1).corr(method="pearson")
feat_scores = corr["__t__"].drop("__t__").abs().sort_values(ascending=False)
selected = feat_scores.index[:TOP_FEATURES].tolist()
X = X[selected]


skews = X.skew().abs()
scaler = RobustScaler() if (skews > 1.0).any() else StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

X_train_full, X_test, y_train_full, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=RANDOM_STATE, stratify=y
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.30, random_state=RANDOM_STATE, stratify=y_train_full
)
print("Shapes: train", X_train.shape, "val", X_val.shape, "test", X_test.shape)

rows = []
for name, p in METRICS.items():
    for k in KS:
        clf = KNeighborsClassifier(n_neighbors=k, p=p)
        clf.fit(X_train, y_train)
        rows.append({
            "metric": name,
            "p": p,
            "k": k,
            "train_acc": accuracy_score(y_train, clf.predict(X_train)),
            "val_acc": accuracy_score(y_val, clf.predict(X_val)),
            "test_acc": accuracy_score(y_test, clf.predict(X_test))
        })
results = pd.DataFrame(rows).sort_values(["metric","k"]).reset_index(drop=True)
print("KNN results:\n", results)

best_row = results.loc[results["val_acc"].idxmax()]
print("\nBest by val:", best_row.to_dict())
best = KNeighborsClassifier(n_neighbors=int(best_row.k), p=int(best_row.p))
best.fit(X_train, y_train)
y_test_pred = best.predict(X_test)
print("Test classification report:\n", classification_report(y_test, y_test_pred, zero_division=0))
print("Confusion matrix:\n", confusion_matrix(y_test, y_test_pred))

print("Outputs saved to", OUTDIR)









```
